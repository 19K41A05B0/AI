```python
import numpy as np
import math
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
eta=0.1
iter_max=100
```

```python
iter=1
while iter<=iter_max :
  slope_x=6*x
  slope_y=-5*(math.exp(-y))
  step_length_x=-eta*slope_x
  step_length_y=-eta*slope_y
  x=x+step_length_x
  y=y+step_length_y
  iter=iter+1
  obj=3*x*x+5*(math.exp(-y))+10
  print("objective function = ",round(obj),"at x=",round(x),",y=",round(y))
```
